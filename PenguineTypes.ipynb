```python
import os
import pandas as pd
import requests

# Create a folder to save the images
folder_name = 'booksImage'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Function to download and save images
def download_images(df):
    for index, row in df.iterrows():
        isbn13 = row['isbn13']
        image_url = row['thumbnail']
        file_path = os.path.join(folder_name, f'{isbn13}.jpg')
        
        # Check if the image file already exists
        if not os.path.exists(file_path):
            if pd.notna(image_url):  # Check if the URL is valid
                response = requests.get(image_url)
                if response.status_code == 200:
                    with open(file_path, 'wb') as file:
                        file.write(response.content)
                    print(f"Downloaded and saved image for ISBN13: {isbn13}")
                else:
                    print(f"Failed to download image for ISBN13: {isbn13}")
            else:
                print(f"Invalid URL for ISBN13: {isbn13}")
        else:
            pass
# Download and save images
download_images(check)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(predictor, target, test_size=0.3)
xdev, xval, ydev, yval=train_test_split(xtest, ytest, test_size=0.15)
```

```python
def preprocess_train(xtrain):
    scaler_list=[]
    for col in range(1,5):
        scaler=MinMaxScaler(feature_range=(0,1000))
        scaler.fit(xtrain.iloc[:,col].values.reshape(-1,1))
        scaler_list.append(scaler)
        xtrain.iloc[:,col]=scaler.transform(xtrain.iloc[:,col].values.reshape(-1,1)).reshape(-1,)
    scaler_yr=MinMaxScaler(feature_range=(0,1000))
    scaler_yr.fit(xtrain.iloc[:,6].values.reshape(-1,1))
    xtrain.iloc[:,6]=scaler_yr.transform(xtrain.iloc[:,6].values.reshape(-1,1)).reshape(-1,)
    return xtrain, scaler_list, scaler_yr

def preprocess_test(xtest, scaler_list, scaler_yr):
    for i,scaler in enumerate(scaler_list, start=1):
        xtest.iloc[:,i]=scaler.transform(xtest.iloc[:,i].values.reshape(-1,1)).reshape(-1,)
    xtest.iloc[:,6]=scaler_yr.transform(xtest.iloc[:,6].values.reshape(-1,1)).reshape(-1,)    
    return xtest
```

```python
xtrain_nm, scaler_list, scaler_yr=preprocess_train(xtrain)
xdev_nm=preprocess_test(xdev, scaler_list, scaler_yr)
plt.subplot(1,2,1)
ax=sns.boxplot(pd.DataFrame(xtrain_nm))
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.subplot(1,2,2)
ax2=sns.boxplot(pd.DataFrame(xdev_nm))
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=90)
plt.show()
```
